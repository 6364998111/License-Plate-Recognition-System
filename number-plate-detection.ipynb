```python
import torch
import os
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image, ImageDraw

# Define class names from your data.yaml
CLASS_NAMES = ["numberplate"]

# Custom dataset for Rail Defect Detection
class RailDefectDataset(Dataset):
    def __init__(self, image_dir, label_dir, transform=None):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.image_files = sorted(os.listdir(image_dir))  # Get all image filenames
        self.transform = transform

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.image_dir, self.image_files[idx])
        image = Image.open(img_path).convert("RGB")

        # Load corresponding label file
        label_path = os.path.join(self.label_dir, self.image_files[idx].replace(".jpg", ".txt"))
        boxes, class_ids = [], []
        if os.path.exists(label_path):
            with open(label_path, "r") as f:
                for line in f.readlines():
                    data = line.strip().split()
                    if len(data) >= 5:  # Ensure we have at least 5 values per line
                        class_id, x_center, y_center, width, height = map(float, data[:5])  # Ignore extra values
                        class_ids.append(int(class_id))
                        boxes.append([x_center, y_center, width, height])

        if self.transform:
            image = self.transform(image)

        return image, boxes, class_ids

# Define transformations
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# Custom collate function to handle variable-sized bounding boxes
def collate_fn(batch):
    images, boxes, class_ids = zip(*batch)  # Unzip the batch
    images = torch.stack(images, dim=0)  # Stack images into a single tensor
    return images, list(boxes), list(class_ids)  # Keep boxes and class_ids as lists

# Load dataset with DataLoader
image_dir = "/content/License-Plate-Recognition-8/valid/images"
label_dir = "/content/License-Plate-Recognition-8/valid/labels"
val_dataset = RailDefectDataset(image_dir, label_dir, transform=transform)
val_loader = DataLoader(val_dataset, batch_size=4, shuffle=True, collate_fn=collate_fn)

# Function to draw multiple bounding boxes
def draw_bounding_boxes(image, boxes, class_ids):
    image = transforms.ToPILImage()(image)  # Convert tensor to PIL image
    draw = ImageDraw.Draw(image)
    width, height = image.size

    for box, class_id in zip(boxes, class_ids):
        x_center, y_center, w, h = box
        x_min = (x_center - w / 2) * width
        y_min = (y_center - h / 2) * height
        x_max = (x_center + w / 2) * width
        y_max = (y_center + h / 2) * height

        # Draw rectangle
        draw.rectangle([x_min, y_min, x_max, y_max], outline="red", width=3)
        # Add class label
        draw.text((x_min, y_min - 10), CLASS_NAMES[class_id], fill="red")

    return image

# Get a batch of data
dataiter = iter(val_loader)
inputs, boxes_batch, class_ids_batch = next(dataiter)

# Plot multiple images with bounding boxes
fig, axes = plt.subplots(1, 4, figsize=(16, 6))
for i in range(4):
    img_with_boxes = draw_bounding_boxes(inputs[i], boxes_batch[i], class_ids_batch[i])
    axes[i].imshow(img_with_boxes)
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```

```python
import gradio as gr
import torch
from ultralytics import YOLO
from PIL import Image, ImageDraw
import numpy as np
import easyocr

# Load YOLO model
model_path = "/content/runs/detect/train/weights/best.pt"  # Update with your model path
model = YOLO(model_path)
CLASS_NAMES = model.names  # Get class names from model

# Initialize EasyOCR Reader
reader = easyocr.Reader(["en"])  # English language support

# Function to make predictions and extract text
def predict(image):
    image = image.convert("RGB")
    results = model(image)  # Run inference
    pred_boxes = results[0].boxes.xyxy.cpu().numpy()
    pred_scores = results[0].boxes.conf.cpu().numpy()
    pred_labels = results[0].boxes.cls.cpu().numpy().astype(int)

    # Draw bounding boxes
    draw = ImageDraw.Draw(image)
    extracted_texts = []

    for box, label, score in zip(pred_boxes, pred_labels, pred_scores):
        x_min, y_min, x_max, y_max = box
        draw.rectangle([x_min, y_min, x_max, y_max], outline="blue", width=3)
        text = f"{CLASS_NAMES[label]}: {score:.2f}"
        draw.text((x_min, y_min - 10), text, fill="blue")

        # Crop the detected region for text extraction
        cropped_region = image.crop((x_min, y_min, x_max, y_max))
        detected_text = reader.readtext(np.array(cropped_region), detail=0)

        if detected_text:
            extracted_texts.append(f"{CLASS_NAMES[label]}: {' '.join(detected_text)}")

    # Join extracted texts for display
    extracted_text_display = "\n".join(extracted_texts) if extracted_texts else "No text detected"

    return image, extracted_text_display

# Create Gradio interface
iface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="pil"),
    outputs=[gr.Image(type="pil"), gr.Textbox()],
    title="YOLOv5 license plate detection with GRU",
    description="Upload an image to detect rail defects using YOLOv8 and extract text from the bounding boxes.",
)

# Launch the app
iface.launch(debug=True)
```
